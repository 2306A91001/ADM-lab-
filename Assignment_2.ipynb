```python

import pandas as pd
```

```python
df = pd.read_csv('/content/employee_performance.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display basic statistics
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=10, kde=True, color='blue')
plt.title("Years_of_Experience")
plt.xlabel("Monthly_Salary")
plt.ylabel("Performance_Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Performance_Score'], y=df['Years_of_Experience'], hue=df['Monthly_Salary'])
plt.title("Years_of_Experience vs. Performance_Score")
plt.xlabel("Performance_Score")
plt.ylabel("Years of Experience")

```

```python
numeric_df = df.select_dtypes(include=['number']) # Selects only numeric columns
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
